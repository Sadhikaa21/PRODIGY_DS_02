```python
data.fillna({'Age':data['Age'].median()})
data.fillna({'Embarked':data['Embarked'].mode()[0]})
data.drop(columns=['Cabin'],inplace=True)
```

```python
data1=data.select_dtypes(include=['float64','int64'])
correlations=data1.corr()
```

```python
plt.figure()
sns.heatmap(correlations,xticklabels=correlations.columns,yticklabels=correlations.columns,annot=True)
plt.title('Correlations heatmap')
plt.show()
```

```python
plt.figure()
sns.countplot(x='Survived',hue='Sex',data=data)
plt.title('Survival count by Sex')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x='Survived', hue='Pclass', data=data)
plt.title('Survival Count by Class')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure()
sns.histplot(data['Age'],kde=True,bins=30)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure()
sns.histplot(data['Fare'],bins=30,kde=True)
plt.title('Distribution of Fare')
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Survived', hue='Sex', data=data)
plt.title('Survival Distribution by Gender')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Survived', hue='Pclass', data=data)
plt.title('Survival Distribution by Passenger Class')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(data['Age'], data['Fare'])
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')
plt.show()
```

```python
sns.boxplot(x='Pclass', y='Age', data=data)
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.title('Box Plot of Age Distribution by Passenger Class')
plt.show()
```

```python
sns.barplot(x='Pclass', y='Fare', data=data)
plt.xlabel('Passenger Class')
plt.ylabel('Average Fare')
plt.title('Bar Plot of Average Fare by Passenger Class')
plt.show()
```

```python
sns.violinplot(x='Survived', y='Age',hue='Pclass', data=data)
plt.xlabel('Survived')
plt.ylabel('Age')
plt.title('Age Distribution by Survival Status')
plt.legend(title='Passenger Class', loc='best')
plt.show()
```
